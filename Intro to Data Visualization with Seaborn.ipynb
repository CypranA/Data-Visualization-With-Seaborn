```python
countries = pd.read_csv('countries-of-the-world.csv')
countries.head()
```

In this exercise, we'll use a dataset that contains information about 227 countries. This dataset has lots of interesting information on each country, such as the country's birth rates, death rates, and its gross domestic product (GDP). GDP is the value of all the goods and services produced in a year, expressed as dollars per person.

We've created three lists of data from this dataset to get you started. gdp is a list that contains the value of GDP per country, expressed as dollars per person. phones is a list of the number of mobile phones per 1,000 people in that country. Finally, percent_literate is a list that contains the percent of each country's population that can read and write.

```python
# Create scatter plot with GDP on the x-axis and number of phones on the y-axis
sns.scatterplot(x='GDP ($ per capita)', y='Phones (per 1000)', data=countries)
plt.show()
```

```python
# Create scatter plot with GDP on the x-axis and number of phones on the y-axis
sns.scatterplot(x=countries['GDP ($ per capita)'], y=countries['Phones (per 1000)'])
plt.show()
```

```python
# Change this scatter plot to have percent literate on the y-axis
sns.scatterplot(x=countries['GDP ($ per capita)'], y=countries['Literacy (%)'])
plt.show()
```

In the last exercise, we explored a dataset that contains information about 227 countries. Let's do more exploration of this data - specifically, how many countries are in each region of the world?

To do this, we'll need to use a count plot. Count plots take in a categorical list and return bars that represent the number of list entries per category. You can create one here using a list of regions for each country, which is a variable named region.

```python
# Create count plot with region on the y-axis
sns.countplot(y=countries['Region'])
plt.show()
```

# Relational Plots [relplot()] and Subplots

Relplot() takes some of the same arguments as scatterplot , such as data, x, y , and hue , but adds other arguments as well. The major difference between relplot and scatterplot (as well as others) is that it can also segregate data attributes into rows and columns of subplots.

```python
#using relplot() instead of scatterplot()

sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter')

#making a scatter plot with relplots()
```

```python
#using relplot() instead of scatterplot()
#subplots in columns

sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', col='smoker')

#col funtion is used to create separate diagrams for values of smoker column (yes/no)
#col function is used to arrange the diagrams in horizontally (column structure)
```

```python
#using relplot() instead of scatterplot()
#subplots in rows

sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', row='smoker')

#row function is used to arrange the diagrams in vertically (column structure)
```

```python
#using relplot() instead of scatterplot()
#subplots in rows and columns

sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', col='smoker', row='time')

#using col and row function at the same time
```

```python
#using relplot() instead of scatterplot()
#too many subplots in column

sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', col='day')

#when the col specified has more than 2 values for example 'day' column has Sun, Sat, Thur, Fri. Entering the function above will have the diagrams displayed in a single row which is hard to read.
#we have to specify how many diagrams to display per row for easy visibility using col_wrap function
```

```python
#using relplot() instead of scatterplot()
#using col_wrap to fix multiple subplots in columns

sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', col='day', col_wrap=2, col_order=['Thur', 
                                                                                                  'Fri', 
                                                                                                  'Sat', 
                                                                                                  'Sun'])
#col_order function is used to arrange the list of values in order
```

# Customizing scatter plots points using replots()

```python
tips = sns.load_dataset('tips')
tips

#size on the tips dataset represents the number of customers per each row
```

```python
#changing plot transparency

sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', alpha=0.6)

#transparency range is between 0 to 1 with 
#0=completely transparent
#1= completely non transparent
```

# Scatter Plot Points Exercise

In this exercise, we'll explore Seaborn's mpg dataset, which contains one row per car model and includes information such as the year the car was made, the number of miles per gallon ("M.P.G.") it achieves, the power of its engine (measured in "horsepower"), and its country of origin.

What is the relationship between the power of a car's engine ("horsepower") and its fuel efficiency ("mpg")? And how does this relationship vary by the number of cylinders ("cylinders") the car has? Let's find out.

Let's continue to use relplot() instead of scatterplot() since it offers more flexibility.

```python
mpg = pd.read_csv('mpg.csv', index_col=0)
mpg
```

```python
# Import Matplotlib and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns

# Create scatter plot of horsepower vs. mpg
sns.relplot(x='horsepower', y='mpg', data=mpg, kind='scatter', size='cylinders', hue='cylinders')

# Show plot
plt.show()
```

Multiple observation per x-value

```python
air_df = pd.read_csv('air_df.csv')
air_df
```

```python
#plotting multiple observation per x value using scatter plot
print(sns.relplot(x='hour', y='NO_2', data=air_df, kind='scatter'))

#plotting multiple observation per x value using line plot
print(sns.relplot(x='hour', y='NO_2', data=air_df, kind='line'))

#python automatically displays the plot of multiple observations by finding the mean of each observation
#the shaded region on the line plot is the confidence level

#replacing confidence level with standard deviation using the ci parameter
print(sns.relplot(x='hour', y='NO_2', data=air_df, kind='line', ci='sd'))

#turning off the confidence level
print(sns.relplot(x='hour', y='NO_2', data=air_df, kind='line', ci=None))
```

# Line Plot Exercise

In this exercise, we'll continue to explore Seaborn's mpg dataset, which contains one row per car model and includes information such as the year the car was made, its fuel efficiency (measured in "miles per gallon" or "M.P.G"), and its country of origin (USA, Europe, or Japan).

How has the average miles per gallon achieved by these cars changed over time? Let's use line plots to find out!

```python
mpg
```

```python
# Import Matplotlib and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns

# Create line plot
sns.relplot(x='model_year', y='mpg', data=mpg, kind='line')


# Show plot
plt.show()
```

```python
# Make the shaded area show the standard deviation
sns.relplot(x="model_year", y="mpg",
            data=mpg, kind="line", ci='sd')

# Show plot
plt.show()
```

# Exercise cont'd

Let's continue to look at the mpg dataset. We've seen that the average miles per gallon for cars has increased over time, but how has the average horsepower for cars changed over time? And does this trend differ by country of origin?

```python
#turn off confidence interval
# Create line plot of model year vs. horsepower
sns.relplot(x='model_year', y='horsepower', data=mpg, kind='line', ci=None)
```

```python
# Change to create subgroups for country of origin
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, style='origin', hue='origin')
```

```python
# Add markers and make each line have the same style
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, style="origin", 
            hue="origin", markers=True, dashes=False)

# Show plot
plt.show()
```
